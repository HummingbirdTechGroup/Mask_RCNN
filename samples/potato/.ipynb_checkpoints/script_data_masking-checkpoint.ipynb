```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

mask_flat =np.asarray([[0,1,1,0],[0,1,1,1],[0,1,1,0]])
plt.imshow(mask_flat)
plt.show()
```

```python
mask = np.dstack((mask_flat,mask_flat))
mask.shape
```

```python
p_xs = [[1,0,0,0],[2,1,1,1]]
p_ys= [[1,2,1,0],[1,0,1,1]]
import skimage
mask_pol = np.zeros([mask.shape[0], mask.shape[1],
                     mask.shape[2]],
                    dtype=np.uint8)
```

```python
for i, (p_x,p_y) in enumerate(zip(p_xs,p_ys)):
    print(p_x,p_y)
    # Get indexes of pixels inside the polygon and set them to 1
    rr, cc = skimage.draw.polygon(p_x, p_y)
    mask_pol[rr, cc, i] = 1
    mask[:,:,i] = mask[:,:,i]* mask_pol[:,:,i]
    
    fig,axs = plt.subplots(1,2,figsize=(16,16))
    axs[0].imshow(mask[:,:,i])
    axs[0].set_title('Mask final')
    axs[1].imshow(mask_pol[:,:,i])
    axs[1].set_title('Bounding box')

mask =  mask.astype(np.uint8)
```
